```python
import pandas as pd
import numpy as np
import pandasql as ps

from pandasql import sqldf
```

```python
teacher_df = pd.read_csv("./sample_data_teacher.csv")
```

```python
teacher_df["day of week"] = teacher_df["day of week"].apply(lambda row: row.split(","))
teacher_df["time available"] = teacher_df["time available"].apply(lambda row: row.split(","))

teacher_df["day_of_week"] = teacher_df["day of week"]
teacher_df["time_available"] = teacher_df["time available"]

teacher_df = teacher_df.explode("day_of_week")
teacher_df = teacher_df.explode("time_available")
```
